```python
#!pip install nltk
import nltk
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('vader_lexicon')

from wordcloud import WordCloud
```

<a id="5b"></a>
<font color="tblack" size=+1.5><b>b. Data Preprocessing</b></font>

**Stopwords removal**

```python
#generate stop words list
stop_words = set(stopwords.words('english'))

#drop stop words in Reviews
df['Reviews_cleaned'] = df['Reviews'].apply(lambda x: [word for word in str(x).split() if word not in stop_words])

#merge and lowercase
df['Reviews_cleaned'] = df['Reviews_cleaned'].str.join(sep=' ')
df['Reviews_cleaned'] = df['Reviews_cleaned'].str.lower()

#display
df['Reviews_cleaned'].head()
```

**Punctuations Removal**

```python
#english punctuation
eng_punc = string.punctuation
list_punc = list(string.punctuation)

#removing punctuation
for punc in list_punc:
  df['Reviews_cleaned'] = df['Reviews_cleaned'].str.replace(punc,'')

#double-check
df[df['Reviews_cleaned'].str.find("'") != -1]
```

**Lemmantization**

```python
#lemmatization
lemmatizer = WordNetLemmatizer()
df['Reviews_cleaned'] = df['Reviews_cleaned'].apply(lambda x :lemmatizer.lemmatize(x))

#display
df['Reviews_cleaned'].head()
```

**Basic Classification to Pos/Neg Reviews**

```python
#generate model
sia = SentimentIntensityAnalyzer()

#generate scores
df['Review_class'] = df['Reviews_cleaned'].apply(lambda x: sia.polarity_scores(x))

#slicing cols
df[['neg','neu','pos','compound']] = df['Review_class'].apply(lambda x: pd.Series(x))

#segment review & drop cols
df['Review_class'] = df['compound'].apply(lambda x: "pos" if x>=0 else "neg")
#df.drop(columns=['neg','neu','pos'], inplace=True)

#display
df['Review_class'].head()
```

**Total Sentiment**

```python
#clean uneccesary words for wordCloud
for word in ['i','phone','']:
  df['Reviews_cleaned'] = df['Reviews_cleaned'].str.replace(word, '')
```

```python
#concat into big string
df_all = df.copy()
df_all['Reviews_cleaned'] = df_all['Reviews_cleaned'].apply(lambda x: [word for word in str(x).split()])
all_words = [word for review in df_all['Reviews_cleaned'].tolist() for word in review]
all_words_str = ' '.join(all_words)

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words_str)

# Display
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#stuff list
stuff_list = ['battery','sm card', 'sd card', 'new', 'touch screen', 'screen protector', 'feature', 'problem']
fig, axes = plt.subplots(nrows=len(stuff_list), ncols=1, figsize=(8, 6 * len(stuff_list)))

for idx, stuff in enumerate(stuff_list):
  #data adjustment
  df_stuff = df[df['Reviews_cleaned'].str.find(stuff) != -1].groupby('Review_class').agg({'Product Name':'count'})
  df_stuff.reset_index(inplace=True)

  #barplot
  sns.barplot(data=df_stuff, x='Review_class', y='Product Name',  ax=axes[idx])
  axes[idx].set_title(f'Bar Plot for {stuff.capitalize()}')
  axes[idx].set_xlabel('Sentiment Class')
  axes[idx].set_ylabel('Number of comments')

  #number above each bar
  for p in axes[idx].patches:
        axes[idx].text(p.get_x() + p.get_width() / 2., p.get_height(), f'{p.get_height()}', ha='center', va='bottom')

#display
plt.show()
```
