```python
#create product_df with its average price
product_df = df[df['Price'].isnull() == False].groupby('Product Name').agg({'Price':'mean'})
product_df['Product Name'] = product_df.index
product_df.reset_index(drop=True, inplace=True)

product_df.head()

#iterrate through rows that has null price value in main df
#replace any null price value with existing price value that product_df
#replace any null price value that does not exist in product_df with total average price in the main df
for index, row in df[df['Price'].isnull()].iterrows():
  for index_2, row_2 in product_df.iterrows():
    if row['Product Name'] == row_2['Product Name']:
      row['Price'] = row_2['Price']
    else:
      row['Price'] = df['Price'].mean()
```

```python
#!pip install nltk
import nltk
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('vader_lexicon')

from wordcloud import WordCloud
```

<a id="5b"></a>
<font color="tblack" size=+1.5><b>b. Data Preprocessing</b></font>

**Stopwords removal**

```python
#generate stop words list
stop_words = set(stopwords.words('english'))

#drop stop words in Reviews
df['Reviews_cleaned'] = df['Reviews'].apply(lambda x: [word for word in str(x).split() if word not in stop_words])

#merge and lowercase
df['Reviews_cleaned'] = df['Reviews_cleaned'].str.join(sep=' ')
df['Reviews_cleaned'] = df['Reviews_cleaned'].str.lower()

#display
df['Reviews_cleaned'].head()
```

**Punctuations Removal**

```python
#english punctuation
eng_punc = string.punctuation

#removing punctuation
def cleaning_punctuations(text):
    translator = str.maketrans('', '', eng_punc)
    return text.translate(translator)

df['Reviews_cleaned'] = df['Reviews_cleaned'].apply(lambda x: cleaning_punctuations(x))

#display
df['Reviews_cleaned'].tail()
```

**Lemmantization**

```python
#lemmatization
lemmatizer = WordNetLemmatizer()
df['Reviews_cleaned'] = df['Reviews_cleaned'].apply(lambda x :lemmatizer.lemmatize(x))

#display
df['Reviews_cleaned'].head()
```

**Basic Classification to Pos/Neg Reviews**

```python
#generate model
sia = SentimentIntensityAnalyzer()

#generate scores
df['Review_class'] = df['Reviews_cleaned'].apply(lambda x: sia.polarity_scores(x))

#slicing cols
df[['neg','neu','pos','compound']] = df['Review_class'].apply(lambda x: pd.Series(x))

#segment review & drop cols
df['Review_class'] = df['compound'].apply(lambda x: "pos" if x>=0 else "neg")
#df.drop(columns=['neg','neu','pos'], inplace=True)

#display
df['Review_class'].head()
```

**Total Sentiment**

```python
#concat into big string
df['Reviews_cleaned'] = df['Reviews_cleaned'].apply(lambda x: [word for word in str(x).split()])
all_words = [word for review in df['Reviews_cleaned'].tolist() for word in review]
all_words_str = ' '.join(all_words)

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words_str)

# Display
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
